## `conll_to_text`

```python
from estnltk.converters.conll.conll_importer import conll_to_text, add_layer_from_conll


text = conll_to_text(file='conll_importer.conll',
                     syntax_layer='conll_syntax_1'  # default: 'conll_syntax'
                     )
text.conll_syntax_1
```

```python
# browse dependency tree
span = text.conll_syntax_1[0]

while span:
    display(span)
    span = span.parent_span
```
