## Rule implementation

### adj_modifier/modifies & predicate_Adj/predicate_Adj_of

```python
collocations = []
forms = [("n", "n"), ("g", "g"), ("p", "p"), ("ill", "ill"), ("in", "in"), ("el", "el"), ("all", "all"), ("ad", "ad"), ("abl", "abl"), ("tr", "tr"), ("g", "ter"), ("g", "es"), ("g", "ab"), ("g", "kom")]

for sentence in text.sentences:
    for i, word1 in enumerate(sentence):
        if len(word1.partofspeech) != 1 or word1.partofspeech[0] != 'A':
            continue
                    
        if i + 1 < len(sentence):
            word2 = sentence[i + 1]
            
            if len(word2.partofspeech) == 1 and word2.partofspeech[0] == 'A':
                if i + 2 < len(sentence):
                    word3 = sentence[i + 2]
                    
                    if len(word3.partofspeech) != 1 or word3.partofspeech[0] != 'S':
                        continue
                    
                    word1_form = word1.form[0]
                    word2_form = word2.form[0]
                    word3_form = word3.form[0]
                    
                    for form in forms:
                        if word1_form.endswith(form[0]) and word2_form.endswith(form[0]) and word3_form.endswith(form[1]):
                            print(word1.lemma[0], word2.lemma[0], word3.lemma[0])
                            print(word1.partofspeech[0], word2.partofspeech[0], word3.partofspeech[0])
                            print(word1_form, word2_form, word3_form)
                            print(form)
                            print("Lause:", " ".join(sentence.text))
                            print("---------")
                            collocations.append((word1.lemma[0], word3.lemma[0]))
                            collocations.append((word2.lemma[0], word3.lemma[0]))
                    break # kui juba leidis adj1 adj2 noun, siis liigu edasi, muidu järgmise ifi juures leiab adj2 noun
                    
            
            if len(word2.partofspeech) == 1 and word2.partofspeech[0] == 'S':
                
                word1_form = word1.form[0]
                word2_form = word2.form[0]

                for form in forms:
                    if word1_form.endswith(form[0]) and word2_form.endswith(form[1]):
                        print(word1.lemma[0], word2.lemma[0])
                        print(word1.partofspeech[0], word2.partofspeech[0])
                        print(word1_form, word2_form)
                        print(form)
                        print("Lause:", " ".join(sentence.text))
                        print("---------")
                        collocations.append((word1.lemma[0], word2.lemma[0]))
                    
        if i - 2 >= 0:
            word3 = sentence[i - 1]
            word2 = sentence[i - 2]
            
            if word3.partofspeech[0] == 'V' and word3.lemma[0] == 'olema' and word2.partofspeech[0] == 'S':
                if i + 1 < len(sentence):
                    word_next = sentence[i + 1]
                    if word_next.partofspeech[0] in ['V', 'S']:
                        continue
                
                word1_form = word1.form[0]
                word2_form = word2.form[0]

                if word1_form.endswith("n") and word2_form.endswith("n"):
                    print(word1.lemma[0], word2.lemma[0])
                    print(word1.partofspeech[0], word2.partofspeech[0])
                    print(word1_form, word2_form)
                    print("Lause:", " ".join(sentence.text))
                    print("---------")
                    collocations.append((word1.lemma[0], word2.lemma[0]))
                        
            if word2.partofspeech[0] == 'V' and word2.lemma[0] == 'olema' and word3.partofspeech[0] == 'S':
                if i + 1 < len(sentence):
                    word_next = sentence[i + 1]
                    if word_next.partofspeech[0] == 'S':
                        continue
                        
                word1_form = word1.form[0]
                word3_form = word3.form[0]
                
                if word1_form.endswith("n") and word3_form.endswith("n"):
                    print(word1.lemma[0], word3.lemma[0])
                    print(word1.partofspeech[0], word3.partofspeech[0])
                    print(word1_form, word3_form)
                    print("Lause:", " ".join(sentence.text))
                    print("---------")
                    collocations.append((word1.lemma[0], word3.lemma[0]))
```

### subject/subject_of

```python
"tallinn".startswith("ta")
```

```python
collocations2 = []
collocations_og = []
collocations_ei = []
# forms = [("n", "n"), ("g", "g"), ("p", "p"), ("ill", "ill"), ("in", "in"), ("el", "el"), ("all", "all"), ("ad", "ad"), ("abl", "abl"), ("tr", "tr"), ("g", "ter"), ("g", "es"), ("g", "ab"), ("g", "kom")]
features_start = ["ta", "ti", "tu"]
features_wrong = ["nud", "da", "tud", "des", "ma", "maks", "mas", "mast", "mata"]
words_wrong = ["alla", "alt", "edasi", "eemale", "esile", "ette", "juurde", "järele", "kaasa", "kinni", "kokku", "kõrvale", "külge", "lahku", "lahti", "ligi", "läbi", "maha", "mööda", "otsa", "peale", "pealt", "ringi", "sisse", "taga", "tagant", "tagasi", "täis", "vahele", "vastu", "välja", "ära", "üle", "üles", "üleval", "ümber"]
features_start_2 = ["ta", "ti", "tu"]
features_wrong_2 = ["nud", "da"]

for sentence in text.sentences:
    for i, word1 in enumerate(sentence):
        if len(word1.partofspeech) == 1 and word1.partofspeech[0] == 'S':
            
            if not word1.form[0].endswith("n"):
                continue

            if i + 1 < len(sentence):
                word2 = sentence[i + 1]
                ei = False

                if word2.partofspeech[0] == 'D':
                    i += 1
                    if i >= len(sentence):
                        break
                    word2 = sentence[i + 1]

                if word2.text == 'ei':
                    i += 1
                    if i >= len(sentence):
                        break
                    word2 = sentence[i + 1]
                    ei = True

                if len(word2.partofspeech) > 1 or word2.partofspeech[0] != 'V' or word2.lemma[0] == 'olema':
                    continue

                for feat in features_start:
                    if word2.form[0].startswith(feat):
                        continue

                for feat in features_wrong:
                    if word2.form[0] == feat:
                        continue

                if i + 2 < len(sentence):
                    word3 = sentence[i + 2]
                    if word3.partofspeech[0] == 'V' and word3.partofspeech[0] == 'X':
                        continue
                    word3_lemma = word3.lemma
                    is_wrong = sum([False if word3_lemma != wrong_word else True for wrong_word in words_wrong])
                    if is_wrong > 0:
                        print((word1.text, word2.text, word3.text))
                        continue

                if ei:
                    collocations_ei.append((word1.text, "ei " + word2.text))
                collocations2.append((word1.lemma[0], word2.lemma[0]))
                collocations_og.append((word1.text, word2.text))
                
        if len(word1.partofspeech) == 1 and word1.partofspeech[0] == 'V' and word1.lemma[0] != 'olema' and word1.lemma[0] != 'ei':
            for feat in features_start_2:
                if word1.form[0].startswith(feat):
                    continue

            for feat in features_wrong_2:
                if word1.form[0] == feat:
                    continue

            if i + 1 < len(sentence):
                word2 = sentence[i + 1]

                if len(word2.partofspeech) > 1 or word2.partofspeech[0] != 'S':
                    continue

                if not word2.form[0].endswith("n"):
                    continue

                if i + 2 < len(sentence):
                    word3 = sentence[i + 2]
                    if word3.partofspeech[0] == 'V' and word3.partofspeech[0] == 'X':
                        continue
                    word3_lemma = word3.lemma
                    is_wrong = sum([False if word3_lemma != wrong_word else True for wrong_word in words_wrong])
                    if is_wrong > 0:
                        print((word1.text, word2.text, word3.text))
                        continue

                collocations2.append((word2.lemma[0], word1.lemma[0]))
                collocations_og.append((word2.text, word1.text))
```

```python
print(collocations2)
print(collocations_og)
print(collocations_ei)
```

### object/object_of

```python
t = Text("Loom leiti üles.")
```

```python
collocations4 = []
collocations4_og = []

for sentence in t.sentences:
    for i, word1 in enumerate(sentence):
        if len(word1.partofspeech) > 1 or word1.partofspeech[0] != 'S' or not word1.form[0].endswith("n"):
            continue
            
        if i + 1 < len(sentence):
            word2 = sentence[i + 1]
            if len(word2.partofspeech) > 1 or word2.partofspeech[0] != 'V' or word2.form[0] not in ['ti', 'takse']:
                continue
            
        collocations4.append((word1.lemma[0], word2.lemma[0]))
        collocations4_og.append((word1.text, word2.text))
```
