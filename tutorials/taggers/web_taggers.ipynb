# Web taggers

To run a web tagger you need a running [EstNLTK web-tagger](https://github.com/estnltk/webtagger-service) service.

```python
from estnltk import Text
```

## VabamorfWebTagger

See also documentation for `VabamorfTagger`.

```python
from estnltk.taggers import VabamorfWebTagger
vabamorph_web_tagger = VabamorfWebTagger(url='http://127.0.0.1:5000/1.6.7beta/tag/morph_analysis')

text = Text('See on lause.')
text.tag_layer(['sentences'])

vabamorph_web_tagger.tag(text)

text.pop_layer('tokens')

text
```

## BertEmbeddingsWebTagger

See also documentation for `BertTagger`.

```python
from estnltk.taggers import BertEmbeddingsWebTagger
bert_embeddings_tagger = BertEmbeddingsWebTagger(url='http://127.0.0.1:5000/1.6.7beta/tag/bert_embeddings')

bert_embeddings_tagger.tag(text)
text.bert_embeddings
```

## SoftmaxEmbTagSumWebTagger

See also documentation for `SoftmaxEmbTagSumTagger`.

```python
from estnltk.taggers import SoftmaxEmbTagSumWebTagger

text = Text("See on lause.")
text.tag_layer(['morph_analysis'])
        
tagger = SoftmaxEmbTagSumWebTagger(url='http://127.0.0.1:5000/1.6.7beta/tag/morph_softmax_emb_tag_sum',
                                   output_layer='morph_softmax_emb_tag_sum')
tagger.tag(text)
text.morph_softmax_emb_tag_sum
```
