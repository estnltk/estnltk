### Finding SVO collocations from the Stanza output

```python
text.sentences[1].stanza_syntax
```

```python
svo = []

for sentence in text.sentences:
    stanza_layer = sentence.stanza_syntax
    verb = None
    subj = None
    obj = None
    for el in stanza_layer:
        if el.deprel == "root" and el.upostag == "V":
            verb = el
            
    if verb is None:
        continue
    
    verb_id = verb.id
    
    for el in stanza_layer:
        if el.head == verb_id:
            if el.deprel == "obj" and el.upostag in ["S", "P", "N"]:
                obj = el
            if el.deprel == "nsubj" and el.upostag in ["S", "P", "N"]:
                subj = el
                
    if None not in [verb, subj, obj]:
        svo.append((subj.lemma, verb.lemma, obj.lemma))
```

### CollocationNet

```python
from subject_verb_collocation_net import SubjectVerbCollocationNet
from verb_object_collocation_net import VerbObjectCollocationNet
```

```python
sv = SubjectVerbCollocationNet()
vo = VerbObjectCollocationNet()
```

```python
preds = {}

for s, v, o in svo:
    try:
        sv_pred = sv.predict_subject_probabilities(v, [s, o])
        vo_pred = vo.predict_object_probabilities(v, [s, o])
        preds[(s, v, o)] = {"sv": sv_pred, "vo": vo_pred}
    except Exception as e: 
        print(e)
```

```python
len(preds)
```

```python
other_way = []
svo_probs = {}

for col in svo:
    try:
        sv_pred = preds[col]["sv"]
        vo_pred = preds[col]["vo"]
        s, v, o = col
        svo_prob = 1
        ovs_prob = 1
        
        for pred in sv_pred:
            if pred[0] == s:
                svo_prob *= pred[1]
            if pred[0] == o:
                ovs_prob *= pred[1]
        
        for pred in vo_pred:
            if pred[0] == s:
                ovs_prob *= pred[1]
            if pred[0] == o:
                svo_prob *= pred[1]
                
        if ovs_prob > svo_prob:
            other_way.append(col)
        
        svo_probs[col] = {"svo": svo_prob, "ovs": ovs_prob}
    except:
        continue
```

```python
len(other_way)
```

```python
for o in other_way:
    print(o, svo_probs[o])
```
